```python
import matplotlib.pyplot as plt

# 讀入檔案
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/

from collections import defaultdict

word_count = defaultdict(int) #有出現error，加上這行自動建立key，給予初始值 0

with open("/content/drive/MyDrive/HW2_hash_practice 檔案夾/hw2_data.txt", "r") as f:
    for line in f:
        word = line.strip()
        if word:  # 避免空行
            word_count[word] += 1# 每多讀一行字，就加1次次數

# 1. 總共有多少個不重複的英文字
unique_words = len(word_count)
print(f"Total unique words: {unique_words}")

# 2. 每一個英文字出現次數為何
print("Word frequencies:")
for word, count in word_count.items():
    print(f"{word}: {count}")

# 3. 畫出直方圖（由多到少）
sorted_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
words = [x[0] for x in sorted_words]
counts = [x[1] for x in sorted_words]

plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Word Frequency Histogram")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("word_histogram.png")
plt.show()
```
